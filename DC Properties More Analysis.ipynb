```python
#importing library

import pandas as pd
import numpy as np

from matplotlib import pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
#reading datasets

df_DC_Properties = pd.read_csv('DC_Properties.csv',parse_dates=True)
#dtype={'NUM_UNITS':int,'AYB': int, 'YR_RMDL':int, 'STORIES': int,'KITCHENS': int,'ZIPCODE':int  }
```

```python
df_DC_Properties=df_DC_Properties.fillna(0)
df_DC_Properties=df_DC_Properties.astype(dtype={'NUM_UNITS':int,'AYB': int, 'YR_RMDL':int, 'STORIES': int,'KITCHENS': int,'ZIPCODE':int  })
```

```python
#drop the coloumn which has no reference

df_DC_Properties=df_DC_Properties.drop('Unnamed: 0',axis=1)
```

```python
df_DC_Properties.describe()
```

```python
meanbath = df_DC_Properties['BATHRM'].mean()
stdbath = df_DC_Properties['BATHRM'].std()
tobath = meanbath + stdbath * 3
botbath = meanbath - stdbath * 3

newdf = df_DC_Properties.copy() #to not mess up the original df
newdf = newdf.drop(newdf[newdf['BATHRM'] > tobath].index)
newdf = newdf.drop(newdf[newdf['BATHRM'] < botbath].index)

newdf.describe()
```

```python
meanhbath = newdf['HF_BATHRM'].mean()
stdhbath = newdf['HF_BATHRM'].std()
tohbath = meanhbath + stdhbath * 3
bothbath = meanhbath - stdhbath * 3
print(tohbath)
print(bothbath)

newdf_1 = newdf.copy() #to not mess up the original df
newdf_1 = newdf_1.drop(newdf[newdf['HF_BATHRM'] > tohbath].index)
newdf_1 = newdf_1.drop(newdf[newdf['HF_BATHRM'] < bothbath].index)

newdf_1.describe()
```
